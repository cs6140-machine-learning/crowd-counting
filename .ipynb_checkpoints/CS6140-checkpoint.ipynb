```python
import torch
import numpy as np
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import matplotlib.pyplot as plt
import pandas as pd
import torchvision.models as models

from tqdm import trange
from torchvision.transforms import v2
from torch.utils.data import Dataset
from sklearn.model_selection import train_test_split
```

```python
class ImageDataset_cs6140(Dataset):
    def __init__(self, label, image, transform=None):
        self.img_labels = label
        self.img = image
        self.transform = transform
    def __len__(self):
        return len(self.img_labels)
    def __getitem__(self, idx):
        image = self.img[idx]
        label = self.img_labels[idx]
        if self.transform:
            image = self.transform(image)
        return image, torch.tensor(label, dtype=torch.float32)

transforms = v2.Compose([
    v2.ToImage(),
    v2.ToDtype(torch.float32, scale=True),
    # Normalize expects float input
    v2.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])
transforms_resize = v2.Compose([
    v2.ToImage(),
    v2.ToDtype(torch.float32, scale=True),
    v2.Resize(size=(224,224), antialias=True),
    # Normalize expects float input
    v2.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])
```

```python
def train_one_epoch(train_loader, model, device, optimizer, 
                    log_interval, epoch, criterion, batch_size):
    model.train()
    losses = []
    counter = []
    index_count = 0
    total_loss = 0.0
    for img, label in train_loader:
        img, label = img.to(device), label.to(device)
        # ------------------
        # Write your implementation here.
        optimizer.zero_grad()
        result = model(img)
        loss = criterion(result, label)
        loss.backward()
        optimizer.step()
        # ------------------
        index_count += 1
        total_loss += loss.item()
        # Record training loss every log_interval and keep counter of total training images seen
        if (0 == (index_count+1) % log_interval):
            losses.append(loss.item())
            counter.append((index_count * batch_size) + img.size(0) + epoch * len(train_loader.dataset))
    return losses, counter
```

```python
def test_one_epoch(test_loader, model, device, criterion):
    model.eval()
    model.to(device)
    test_loss = 0.0
    num_correct = 0
    count = 0
    with torch.no_grad():
        for img, label in test_loader:
            img, label = img.to(device), label.to(device)
            # ------------------
            # Write your implementation here.
            result = model(img)
            loss = criterion(result, label)
            predicted = torch.round(result)
            num_correct += (predicted == label).sum().item()
            test_loss += loss.item()
            count += 1
            # ------------------
    test_loss /= count
    # test_loss /= len(test_loader.dataset)
    return test_loss, num_correct
```

```python
def plt_loss_graph(trainloader, train_counter, train_losses, test_losses):

    fig = plt.figure(figsize=(12,8))
    plt.plot(train_counter, train_losses, label='Train loss')
    plt.plot([i * len(trainloader.dataset) for i in range(1, len(test_losses) + 1)],
             test_losses, label='Test loss', marker='o')
    plt.xlim(left=0)
    plt.ylim(bottom=0)
    plt.title('Loss curve', fontsize=24)
    plt.xlabel('Number of training examples seen', fontsize=16)
    plt.ylabel('Loss', fontsize=16)
    plt.legend(loc='upper right', fontsize=14)
```

```python
def model_train(trainloader, testloader, model, device,
                criterion, optimizer, max_epochs, batch_size):
    # Recording data
    log_interval = 100

    train_losses = []
    train_counter = []
    test_losses = []
    test_correct = []
    for epoch in trange(max_epochs, leave=True, desc='Epochs'):
        train_loss, counter = train_one_epoch(trainloader, model, device,
                                              optimizer, log_interval, epoch, criterion, batch_size)
        test_loss, num_correct = test_one_epoch(testloader, model, device, criterion)
        # Record results
        train_losses.extend(train_loss)
        train_counter.extend(counter)
        test_losses.append(test_loss)
        test_correct.append(num_correct)

    print()
    print(f"Test accuracy: {test_correct[-1]/len(testloader.dataset)}")

    return train_losses, train_counter, test_losses, test_correct
```

```python
def add_one_to_one_correlation_line(ax, min_factor=1, max_factor=1, **plot_kwargs):
    lim_min, lim_max = pd.DataFrame([ax.get_ylim(), ax.get_xlim()]).agg({0: 'min', 1: 'max'})
    lim_min *= min_factor
    lim_max *= max_factor
    plot_kwargs_internal = dict(color='grey', ls='--')
    plot_kwargs_internal.update(plot_kwargs)
    ax.plot([lim_min, lim_max], [lim_min, lim_max], **plot_kwargs_internal)
    ax.set_ylim([lim_min, lim_max])
    ax.set_xlim([lim_min, lim_max])
```

```python
def model_eval(model, testloader, device):
    model.to(device)
    model.eval()
    all_labels = np.array([])
    all_pred = np.array([])

    with torch.no_grad():
        for img, label in testloader:
          img, label = img.to(device), label.to(device)
          result = model(img)
          result = np.array(result.squeeze(1).tolist())
          all_pred = np.concatenate((all_pred, result), axis=0)
          label = np.array(label.squeeze(1).tolist())
          all_labels = np.concatenate((all_labels, label), axis=0)

    df_predictions = pd.DataFrame({'True values': all_labels, 'Predicted values': all_pred})
    ax = df_predictions.plot.scatter('True values', 'Predicted values', alpha=0.5, s=14, figsize=(9,9))
    ax.grid(axis='both')
    add_one_to_one_correlation_line(ax)
    ax.set_title('Validation')

    plt.show()
```

```python
X = np.load("images.npy")
y = np.load("labels.npy")
#X = np.load("/content/drive/MyDrive/Colab Notebooks/images.npy")
#y = np.load("/content/drive/MyDrive/Colab Notebooks/labels.npy")

# split into 90% train and 10% test
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.1, random_state=514)
# use cuda/gpu
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

Simple CNN model

```python
trainset = ImageDataset_cs6140(y_train, X_train, transform=transforms)
testset = ImageDataset_cs6140(y_test, X_test, transform=transforms)

# setup dataloader
batch_size = 5
trainloader = torch.utils.data.DataLoader(trainset, batch_size=batch_size,
                                          shuffle=True, num_workers=2)
testloader = torch.utils.data.DataLoader(testset, batch_size=batch_size,
                                         shuffle=False, num_workers=2)
```

```python
class Simple_CNN(nn.Module):
    def __init__(self):
        super().__init__()

        self.conv_layer = nn.Sequential(
            nn.Conv2d(3, 16, kernel_size=3, padding=1),
            nn.MaxPool2d(kernel_size=8, stride=8),
            nn.ReLU(True),
            nn.Conv2d(16, 32, kernel_size=3, padding=1),
            nn.MaxPool2d(kernel_size=4, stride=4),
            nn.ReLU(True),
            nn.Flatten()
        )
        self.fc_layer = nn.Sequential(
            nn.Linear(32 * 15 * 20, 1024),
            nn.ReLU(True),
            nn.Linear(1024, 1)
        )

    def forward(self, x):
        x = self.conv_layer(x)
        x = self.fc_layer(x)
        return x
```

```python
model = Simple_CNN()
model.to(device)
```

```python
criterion = nn.SmoothL1Loss()
optimizer = optim.Adam(model.parameters(), lr=0.003)
max_epochs = 50
```

```python
train_losses, train_counter, test_losses, test_correct = model_train(trainloader,
                                                                     testloader, model, device,criterion, 
                                                                     optimizer, max_epochs, batch_size)
```

```python
plt_loss_graph(trainloader, train_counter, train_losses, test_losses)
model_eval(model, testloader, device)
```

Change dataloader, mobile net and vgg16 need 224x224 image input

```python
trainset = ImageDataset_cs6140(y_train, X_train, transform=transforms_resize)
testset = ImageDataset_cs6140(y_test, X_test, transform=transforms_resize)
# setup dataloader
batch_size = 10
trainloader = torch.utils.data.DataLoader(trainset, batch_size=batch_size,
                                          shuffle=True, num_workers=2)
testloader = torch.utils.data.DataLoader(testset, batch_size=batch_size,
                                         shuffle=False, num_workers=2)
```

Mobile net v3 small

```python
model_mobile_net = models.mobilenet_v3_small(weights=models.MobileNet_V3_Small_Weights.IMAGENET1K_V1)
model_mobile_net.classifier[3] = nn.Linear(model_mobile_net.classifier[3].in_features, 1)
```

```python
model = model_mobile_net
model.to(device)
```

```python
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.002)
max_epochs = 70
```

```python
train_losses, train_counter, test_losses, test_correct = model_train(trainloader,testloader, 
                                                                     model, device,criterion, 
                                                                     optimizer, max_epochs, batch_size)
```

```python
plt_loss_graph(trainloader, train_counter, train_losses, test_losses)
model_eval(model, testloader, device)
```

Vgg 16

```python
vgg16 = models.vgg16(weights=models.VGG16_Weights.IMAGENET1K_V1)
vgg16.classifier[-1] = nn.Linear(vgg16.classifier[-1].in_features, 1)
```

```python
model = vgg16
model.to(device)
```

```python
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
max_epochs = 60
```

```python
train_losses, train_counter, test_losses, test_correct = model_train(trainloader,testloader, 
                                                                     model, device,criterion, 
                                                                     optimizer, max_epochs, batch_size)
```

```python
plt_loss_graph(trainloader, train_counter, train_losses, test_losses)
model_eval(model, testloader, device)
```
